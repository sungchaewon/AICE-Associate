```python
# Numpy 라이브러리 불러오기
import numpy as np
```

```python
# 1차원 배열 생성
array = np.array([1, 2, 3, 4, 5])
print("1차원 배열 : ", array)
```

```python
# 배열의 자료형 확안
print("배열의 자료형 : ", array.dtype)
```

```python
# 2차원 배열 생성
matrix = np.array([[1, 2, 3], [4, 5, 6]]) # 2행 3열의 배열 생성
print("2차원 배열 : \n" , matrix)
```

```python
# 배열의 크기 확인
print("배열의 크기 : ", matrix.size) # 배열의 총 요소 개수
# 배열의 형태 확인
print("배열의 형태 : ", matrix.shape) # 배열의 행과 열의 개수
# 배열의 차원 확인
print("배열의 차원 : ", matrix.ndim) # 배열의 차원을 출력
```

```python
# 배열의 합과 평균
print("배열의 합 : ", np.sum(matrix)) # 모든 요소의 합
print("배열의 평균 : ", np.mean(matrix)) # 모든 요소의 평균

# 배열의 최댓값과 최솟값
print("배열의 최댓값 : ", np.max(matrix)) # 배열에서 가장 큰 값
print("배열의 최솟값 : ", np.min(matrix)) # 배열에서 가장 작은 값

print("배열 변경 : ", matrix.reshape(3, 2)) # 배열의 크기를 3개의 행과 2개의 열로 변경
# 배열의 크기(총 요소의 개수)는 변경되지 않아야 함
```

```python
# 배열의 전치
transposed = matrix.T # 행과 열을 바꿈
print("전치된 배열 : \n", transposed)
```

```python
# 추가 파이썬 코드
# 기본 공백 대신 다른 구분자 사용
print("apple", "banana", "cherry", sep=", ")

# 여러 값을 출력해보기
name = "Charlie"
age = "28"
hobby = "reading"

print("이름 : ", name, "나이 : ", age, "취미 : ", hobby)
print(f"이름 : {name}, 나이 : {age}, 취미 : {hobby}")
print("이름 : {}, 나이 : {}, 취미 : {}".format(name, age, hobby))
```
